Each monkey has several attributes:

- `Starting items` lists your **worry level** for each item the monkey is currently holding in the order they will be inspected.
- `Operation` shows how your worry level changes as that monkey inspects an item. (An operation like `new = old * 5` means that your worry level after the monkey inspected the item is five times whatever your worry level was before inspection.)
- `Test` shows how the monkey uses your worry level to decide where to throw an item next.
    - `If true` shows what happens with an item if the `Test` was true.
    - `If false` shows what happens with an item if the `Test` was false.
    
After each monkey inspects an item but before it tests your worry level, your relief that the monkey's inspection didn't damage the item causes your worry level to be **divided by three** and rounded down to the nearest integer.

The monkeys take turns inspecting and throwing items. On a single monkey's **turn**, it inspects and throws all of the items it is holding one at a time and in the order listed. Monkey `0` goes first, then monkey `1`, and so on until each monkey has had one turn. The process of each monkey taking a single turn is called a **round**.

When a monkey throws an item to another monkey, the item goes on the **end** of the recipient monkey's list. A monkey that starts a round with no items could end up inspecting and throwing many items by the time its turn comes around. If a monkey is holding no items at the start of its turn, its turn ends.

Figure out which monkeys to chase by counting how many items they inspect over `20` rounds. What is the level of monkey business after 20 rounds of stuff-slinging simian shenanigans?

````python
# Data Load
example_lines = [
    "Monkey 0:",
    "  Starting items: 79, 98",
    "  Operation: new = old * 19",
    "  Test: divisible by 23",
    "    If true: throw to monkey 2",
    "    If false: throw to monkey 3",
    "",
    "Monkey 1:",
    "  Starting items: 54, 65, 75, 74",
    "  Operation: new = old + 6",
    "  Test: divisible by 19",
    "    If true: throw to monkey 2",
    "    If false: throw to monkey 0",
    "",
    "Monkey 2:",
    "  Starting items: 79, 60, 97",
    "  Operation: new = old * old",
    "  Test: divisible by 13",
    "    If true: throw to monkey 1",
    "    If false: throw to monkey 3",
    "",
    "Monkey 3:",
    "  Starting items: 74",
    "  Operation: new = old + 3",
    "  Test: divisible by 17",
    "    If true: throw to monkey 0",
    "    If false: throw to monkey 1",
]

with open("day_11_input.txt") as _file:
    data_lines = _file.readlines()
````

````python
# Common Code
from __future__ import annotations

class ValueOperation:
    _OPERANDS = {
        "+": lambda x, y: x + y,
        "*": lambda x, y: x * y,
    }

    def __init__(self, operand: str, right: str):
        if operand not in self._OPERANDS:
            raise RuntimeError(f"Unexpected operand: {operand}")

        self._operand = self._OPERANDS[operand]

        self._use_old = right == "old"

        self._right = None if self._use_old else int(right)

    @classmethod
    def parse(cls, line):
        return ValueOperation(*line.split("=")[1].split()[1:])

    def apply(self, old_value):
        right = old_value if self._use_old else self._right
        return self._operand(old_value, right)


class Monkey:
    def __init__(self, items: list[int], operation: ValueOperation, divisible_by, true_throw, false_throw, relief_factor):
        self._items = items
        self._operation = operation
        self._divisible_by = divisible_by
        self._true_throw = true_throw
        self._false_throw = false_throw
        self._relief_factor = relief_factor

        self._throw_count = 0
        
        self.upper_bound = None

    @classmethod
    def parse_monkeys(cls, lines, relief_factor=3):
        monkeys = []

        def _get_int_from_line(index, prefix):
            return int(lines[index].split(prefix)[1].strip())

        # We're going to heavily assume the format doesn't change because:
        # I've checked my input, and I'm lazy
        i = 1
        while i < len(lines):
            items = [int(x.strip()) for x in lines[i].split(":")[1].split(",")]
            operation = ValueOperation.parse(lines[i + 1])
            divisible_by = _get_int_from_line(i + 2, "Test: divisible by")
            true_throw = _get_int_from_line(i + 3, "If true: throw to monkey")
            false_throw = _get_int_from_line(i + 4, "If false: throw to monkey")

            monkeys.append(Monkey(items, operation, divisible_by, true_throw, false_throw, relief_factor))

            i += 7

        return monkeys
    
    @property
    def divisible_by(self):
        return self._divisible_by

    @property
    def items(self):
        return [str(x) for x in self._items]

    @property
    def throw_count(self):
        return self._throw_count

    def _take_turn(self, monkeys: list[Monkey]):
        item = self._items.pop()

        worry_value = self._operation.apply(item)
        worry_value = worry_value // self._relief_factor
        
        if self.upper_bound:
            worry_value %= self.upper_bound

        target = self._true_throw if worry_value % self._divisible_by == 0 else self._false_throw
        target = monkeys[target]

        target._items.append(worry_value)
        self._throw_count += 1

    def take_round(self, monkeys: list[Monkey]):
        while self._items:
            self._take_turn(monkeys)
````

````python
# Solution 1 Code
def solution_1(lines, display=True):
    monkeys = Monkey.parse_monkeys(lines)

    for i in range(20):
        for monkey in monkeys:
            monkey.take_round(monkeys)

        if display:
            print(f"Round {i + 1}")
            for j, monkey in enumerate(monkeys):
                print(f"Monkey {j}: {', '.join(monkey.items)}")
            print()

    throw_counts = [x.throw_count for x in monkeys]
    throw_counts.sort(reverse=True)

    return throw_counts[0] * throw_counts[1]

def test_solution_1():
    actual = solution_1(example_lines)
    expected = 10605
    
    print(actual, expected)
    assert actual == expected
test_solution_1()
````

Worry levels are no longer divided by three after each item is inspected; you'll need to find another way to keep your worry levels manageable. Starting again from the initial state in your puzzle input, **what is the level of monkey business after 10000 rounds**?

````python
import math

def solution_2(lines):
    iterations_to_inspect = [1, 20] + [x * 1000 for x in range(1, 11)]
    
    monkeys = Monkey.parse_monkeys(lines, 1)
    
    upper_bound = math.lcm(*[x.divisible_by for x in monkeys])
    for monkey in monkeys:
        monkey.upper_bound = upper_bound
    
    for i in range(1, 10001):
        for monkey in monkeys:
            monkey.take_round(monkeys)

        if i in iterations_to_inspect:
            print(f"Round {i}")
            for j, monkey in enumerate(monkeys):
                print(f"Monkey {j} inspected items {monkey.throw_count} times")
            print()
        
    throw_counts = [x.throw_count for x in monkeys]
    throw_counts.sort(reverse=True)

    return throw_counts[0] * throw_counts[1]
        
def test_solution_2():
    actual = solution_2(example_lines)
    expected = 2713310158
    
    print(actual, expected)
    assert actual == expected
test_solution_2()
````
